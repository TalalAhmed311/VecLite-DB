```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("AAAMLP-569to.pdf")
pages = loader.load_and_split()
```

```python
from veclite.tokenizer import Tokenizer
tokenizer = Tokenizer()

vectors = []

for page in pages:
    vectors.append(tokenizer.create_embedding(page.page_content))
```

```python
from veclite.vector_store_schema import VectorParams



def create_vector_params(vectors,pages):
    vector_data = []

    for vector, page in zip(vectors,pages):

        vector_data.append(
            VectorParams(vector=vector,content=page.page_content,metadata=page.metadata)
        )

    return vector_data
```

## Create Tree

```python
from veclite.vector_store import VectorStore

store = VectorStore(vector_dim=384,build_tree=True)
```

```python
vector_data = create_vector_params(vectors[:50],pages[:50])
```

```python
store.add_vectors(vector_params=vector_data)
```

```python
store.tree
```

## Search from Tree

```python
vec = vectors[80]

results = store.search_vectors_from_tree(input_vector=vec)
```
